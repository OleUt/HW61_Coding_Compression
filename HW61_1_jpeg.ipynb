```python
import numpy as np
from skimage import io
from sklearn.cluster import KMeans
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# DCT

img = cv2.imread('C:/Users/MyWork/PycharmProjects/HW60_Edge_Filter/dog1.jpg', cv2.IMREAD_GRAYSCALE)

dct = cv2.dct(np.float32(img))
idct = cv2.idct(dct)

plt.imshow(dct)
plt.show()

plt.imshow(idct)
plt.show()

cv2.imwrite('image_dct.jpg', dct)
cv2.imwrite('image_idct.jpg', idct)
```

```python
# quantization

img = io.imread('C:/Users/MyWork/PycharmProjects/HW60_Edge_Filter/dog1.jpg')
pixels = img.reshape((-1, 3))

kmeans = KMeans(n_clusters=16, random_state=0).fit(pixels)
quantized_pixels = kmeans.cluster_centers_[kmeans.labels_]
quantized_img = quantized_pixels.reshape(img.shape)

plt.imshow((quantized_img).astype(np.uint8))
plt.show()

cv2.imwrite('image_quantized.jpg', quantized_img)
```
